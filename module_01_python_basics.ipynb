# Q1: Temperature Conversion

```python

# Get temperature in Celsius from the user. We use float() for decimals.
celsius_temp = float(input("Enter temperature in Celsius: "))

# 1. Convert Celsius to Fahrenheit
# Formula: F = (C * 9/5) + 32
fahrenheit_temp = (celsius_temp * 9/5) + 32

# 2. Convert Celsius to Kelvin
# Formula: K = C + 273.15
kelvin_temp = celsius_temp + 273.15

# Display all results, rounded to two decimal places as required.
print("\n--- Temperature Conversion ---")
print(f"Celsius: {round(celsius_temp, 2)}")
print(f"Fahrenheit: {round(fahrenheit_temp, 2)}")
print(f"Kelvin: {round(kelvin_temp, 2)}\n")
```

# Q2: Number Analysis

```python

num = -7

if num > 0:
    sign = "positive"
elif num < 0:
    sign = "negative"
else:
    sign = "zero"

parity = "even" if num % 2 == 0 else "odd"

print(f"The number is {sign} and {parity}.")
```

# Q3: Conditional Sum

```python
# Q3: Conditional Sum

total = 0

for i in range(1, 101):
    if i % 15 == 0:
        continue
    if i % 3 == 0 or i % 5 == 0:
        total += i
    if total > 2000:
        break

print("Final Sum:", total)
```

# Q1: Student class with constructors

```python
class Student:

    def __init__(self, name=None, student_id=None):
        if name is None and student_id is None:
            self.name = "Unknown"
            self.student_id = 0
        else:
            self.name = name
            self.student_id = student_id

    def display(self):
        print(f"Name: {self.name}, ID: {self.student_id}")

    def placeholder_method(self):
        pass  # reserved for future use


# Testing
s1 = Student()
s2 = Student("Fahim", 101)

s1.display()
s2.display()
```

# Q2: Inheritance examples

```python

class Person:
    def __init__(self, name):
        self.name = name

    def role(self):
        return "Person"


# Single inheritance
class Student(Person):
    def role(self):
        return "Student"


# Multilevel inheritance
class GraduateStudent(Student):
    def role(self):
        return "Graduate Student"


# Hierarchical inheritance
class Teacher(Person):
    def role(self):
        return "Teacher"


# Testing
p = Person("Alex")
s = Student("Fahim")
g = GraduateStudent("Rahim")
t = Teacher("Karim")

print(p.role())
print(s.role())
print(g.role())
print(t.role())
```

# Q3: Encapsulation + Overloading + Multiple Inheritance

```python
class BankAccount:

    def __init__(self, balance):
        self.__balance = balance  # private

    def get_balance(self):
        return self.__balance

    def set_balance(self, amount):
        if amount >= 0:
            self.__balance = amount

    # Simulated method overloading
    def deposit(self, amount, bonus=0):
        self.__balance += amount + bonus


# Multiple inheritance example
class Logger:
    def log(self):
        print("Transaction logged")


class SecureAccount(BankAccount, Logger):
    def show_details(self):
        print(f"Balance: {self.get_balance()}")
        self.log()


# Testing
acc = SecureAccount(1000)
acc.deposit(500)
acc.deposit(200, 50)

acc.show_details()
```

### Insights

- Designed classes with both default and parameterized initialization  
- Used inheritance to model real-world relationships  
- Applied method overriding for behavior customization  
- Implemented encapsulation using private attributes and controlled access  
- Simulated method overloading using default arguments  
- Combined multiple inheritance for extended functionality

### Overview

This notebook explores Python data structures and functions by solving problems involving:

- Removing duplicates and analyzing list data  
- Performing set operations (union, intersection, difference)  
- Working with tuples and understanding their properties

# Q1: Remove duplicates and analyze list

```python
def analyze_numbers(numbers):
    unique = []

    for num in numbers:
        if num not in unique:
            unique.append(num)

    result = {
        "original_list": numbers,
        "unique_values": unique,
        "count": len(unique)
    }

    return result


nums = [1, 2, 2, 3, 4, 4, 5]

output = analyze_numbers(nums)
print(output)
```

# Q2: Set operations

```python
def set_operations(set1=None, set2=None):
    if set1 is None:
        set1 = set()
    if set2 is None:
        set2 = set()

    return {
        "union": set1.union(set2),
        "intersection": set1.intersection(set2),
        "difference": set1.difference(set2)
    }


a = {1, 2, 3}
b = {3, 4, 5}

result = set_operations(a, b)

for key, value in result.items():
    print(f"{key}: {value}")
```

# Q3: Tuple operations

```python
tuple1 = (10, "data", 3.14)
a, b, c = tuple1  # unpacking

tuple2 = (5, "data", 2.71)

print("Tuple 1:", tuple1)
print("Tuple 2:", tuple2)
print("Are they equal?", tuple1 == tuple2)

# Difference:
# Lists are mutable (can be changed after creation)
# Tuples are immutable (cannot be modified once created)
```
